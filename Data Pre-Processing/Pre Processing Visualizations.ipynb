```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# importing the dataset as a csv
df = pd.read_csv(r"E:\UoB\Semester 1\DAPM\Assignments\Assignment 2\Final Dataset CSV\Final Dataset Assignment 2.csv")
```

```python

# Compute the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(
    correlation_matrix, 
    annot=True,  # Show the correlation coefficients
    cmap='coolwarm',  # Choose a color palette
    fmt='.2f',  # Format the correlation coefficients
    square=True,  # Ensure cells are square
    linewidths=0.5,  # Add space between cells
    cbar_kws={"shrink": 0.8}  # Shrink the color bar
)

# Add a title
plt.title('Correlation Heatmap', fontsize=16)

# Show the heatmap
plt.show()
```

```python
# Create multiple histograms
num_columns = df.select_dtypes(include=['number']).columns  # Select numeric columns
num_features = len(num_columns)

# Set the layout of the subplots
rows = (num_features + 1) // 2  # Adjust rows for 2 histograms per row
cols = 2  # Two histograms per row

plt.figure(figsize=(12, rows * 4))  # Adjust figure size based on number of rows

# Create a histogram for each feature
for i, column in enumerate(num_columns, 1):  # Enumerate starting at 1 for subplot indexing
    plt.subplot(rows, cols, i)
    sns.histplot(df[column], kde=True, bins=20, color='blue')  # Histogram with density curve
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# List of columns to exclude
exclude_columns = ['FAF','FCVC']

# Select numeric columns except the excluded ones
selected_columns = [col for col in df.select_dtypes(include=['number']).columns if col not in exclude_columns]

# Create multiple histograms
num_features = len(selected_columns)

# Set the layout of the subplots
rows = (num_features + 1) // 2  # Adjust rows for 2 histograms per row
cols = 2  # Two histograms per row

plt.figure(figsize=(12, rows * 4))  # Adjust figure size based on number of rows

# Create a histogram for each feature
for i, column in enumerate(selected_columns, 1):  # Enumerate starting at 1 for subplot indexing
    plt.subplot(rows, cols, i)
    sns.histplot(df[column], kde=True, bins=20, color='blue')  # Histogram with density curve
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
