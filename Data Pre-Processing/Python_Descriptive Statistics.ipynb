```python
import numpy as np
import pandas as pd
```

```python
# Load data from your CSV file (replace '[your_file].csv' with your actual file path in your project)
df = pd.read_csv("Final Dataset Assignment 2.csv")
df
```

```python
# Calculate descriptive statistics
descriptive_stats = df.describe()
descriptive_stats
```

```python
# Add additional statistics as I have mentioned during lectures, skewness, etc are important
descriptive_stats.loc['variance'] = df.var()
descriptive_stats.loc['skewness'] = df.skew()
descriptive_stats.loc['kurtosis'] = df.kurtosis()
descriptive_stats.loc['median'] = df.median()
# Display the descriptive statistics here in python
print(descriptive_stats)
```
