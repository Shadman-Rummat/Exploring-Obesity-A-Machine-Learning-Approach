```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cdist

# Sample data
data = {
    "Gender": [1, 0, 1, 0],  # Encoded: Male=1, Female=0
    "Age": [41, 41, 16, 17],
    "Height": [1.75, 1.54, 1.82, 1.6],
    "Weight": [118, 80, 71, 53],
    "Meals": [3, 3, 3, 3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute Euclidean distances
euclidean_distances = cdist(df, df, metric='euclidean')

# Compute Manhattan distances
manhattan_distances = cdist(df, df, metric='cityblock')

# Display results
print("Euclidean Distances:\n", euclidean_distances)
print("\nManhattan Distances:\n", manhattan_distances)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import cdist

# Sample data
data = {
    "Gender": ["Male", "Female", "Male", "Female"],  # Retained for labeling
    "Age": [41, 41, 16, 17],
    "Height": [1.8, 1.54, 1.8, 1.6],
    "Weight": [99, 80, 58, 53],
    "Meals": [3, 3, 3, 3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Retain Gender and Age for labeling
labels = df[["Gender", "Age"]]

# Exclude Gender from analysis and standardize
features = df.drop(columns=["Gender"])
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Compute Euclidean and Manhattan distances
euclidean_distances = cdist(features_scaled, features_scaled, metric='euclidean')
manhattan_distances = cdist(features_scaled, features_scaled, metric='cityblock')

# Convert distances to DataFrame for better readability
euclidean_df = pd.DataFrame(euclidean_distances, index=labels.apply(tuple, axis=1), columns=labels.apply(tuple, axis=1))
manhattan_df = pd.DataFrame(manhattan_distances, index=labels.apply(tuple, axis=1), columns=labels.apply(tuple, axis=1))

# Display results
print("Euclidean Distances (Standardized):")
print(euclidean_df)

print("\nManhattan Distances (Standardized):")
print(manhattan_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data for Euclidean distances
euclidean_data = {
    "(Male, 41)": [0.000000, 2.452295, 2.997404, 3.633087],
    "(Female, 41)": [2.452295, 0.000000, 3.221562, 2.534730],
    "(Male, 17)": [2.997404, 3.221562, 0.000000, 1.731813],
    "(Female, 17)": [3.633087, 2.534730, 1.731813, 0.000000],
}
euclidean_df = pd.DataFrame(euclidean_data, index=["(Male, 41)", "(Female, 41)", "(Male, 17)", "(Female, 17)"])

# Data for Manhattan distances
manhattan_data = {
    "(Male, 41)": [0.000000, 3.257972, 4.232584, 6.215127],
    "(Female, 41)": [3.257972, 0.000000, 5.421332, 3.983321],
    "(Male, 17)": [4.232584, 5.421332, 0.000000, 1.982543],
    "(Female, 17)": [6.215127, 3.983321, 1.982543, 0.000000],
}
manhattan_df = pd.DataFrame(manhattan_data, index=["(Male, 41)", "(Female, 41)", "(Male, 17)", "(Female, 17)"])

# Plotting the heatmaps
plt.figure(figsize=(12, 6))

# Euclidean distance heatmap
plt.subplot(1, 2, 1)
sns.heatmap(euclidean_df, annot=True, cmap="Blues", fmt=".3f", cbar_kws={'label': 'Distance'})
plt.title("Euclidean Distances (Standardized)")
plt.xlabel("Individual")
plt.ylabel("Individual")

# Manhattan distance heatmap
plt.subplot(1, 2, 2)
sns.heatmap(manhattan_df, annot=True, cmap="Oranges", fmt=".3f", cbar_kws={'label': 'Distance'})
plt.title("Manhattan Distances (Standardized)")
plt.xlabel("Individual")
plt.ylabel("Individual")

# Show the plots
plt.tight_layout()
plt.show()
```
