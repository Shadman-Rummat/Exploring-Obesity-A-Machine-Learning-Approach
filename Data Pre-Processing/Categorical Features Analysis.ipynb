```python
import matplotlib.pyplot as plt

# Data for pie charts
labels_gender = ['Female', 'Male']
sizes_gender = [1043, 1067]
colors_gender = ['#66c2a5', '#fc8d62']

labels_smoker = ['No', 'Yes']
sizes_smoker = [2066, 44]
colors_smoker = ['#8da0cb', '#e78ac3']

labels_family_history = ['No', 'Yes']
sizes_family_history = [385, 1725]
colors_family_history = ['#a6d854', '#ffd92f']

# Create subplots for pie charts
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Gender Pie Chart
axes[0].pie(sizes_gender, labels=labels_gender, autopct='%1.1f%%', startangle=90, colors=colors_gender)
axes[0].set_title('Gender Distribution', fontsize=14)

# Smoker Pie Chart
axes[1].pie(sizes_smoker, labels=labels_smoker, autopct='%1.1f%%', startangle=90, colors=colors_smoker)
axes[1].set_title('Smoking Status', fontsize=14)

# Family History Pie Chart
axes[2].pie(sizes_family_history, labels=labels_family_history, autopct='%1.1f%%', startangle=90, colors=colors_family_history)
axes[2].set_title('Family History of Obesity', fontsize=14)

# Adjust spacing and show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for pie charts
labels_gender = ['Female', 'Male']
sizes_gender = [1043, 1067]
colors_gender = ['#66c2a5', '#fc8d62']

labels_smoker = ['No', 'Yes']
sizes_smoker = [2066, 44]
colors_smoker = ['#8da0cb', '#e78ac3']

labels_family_history = ['No', 'Yes']
sizes_family_history = [385, 1725]
colors_family_history = ['#a6d854', '#ffd92f']

labels_alcohol = ['No', 'Sometimes', 'Frequently']
sizes_alcohol = [639, 1400, 71]
colors_alcohol = ['#66c2a5', '#fc8d62', '#8da0cb']

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('Distribution of Categorical Variables', fontsize=16)

# Gender Pie Chart
axes[0, 0].pie(sizes_gender, labels=labels_gender, autopct='%1.1f%%', colors=colors_gender, startangle=90)
axes[0, 0].set_title('Gender Distribution')

# Smoker Pie Chart
axes[0, 1].pie(sizes_smoker, labels=labels_smoker, autopct='%1.1f%%', colors=colors_smoker, startangle=90)
axes[0, 1].set_title('Smoker Distribution')

# Family History of Obesity Pie Chart
axes[1, 0].pie(sizes_family_history, labels=labels_family_history, autopct='%1.1f%%', colors=colors_family_history, startangle=90)
axes[1, 0].set_title('Family History of Obesity')

# Alcohol Consumption Pie Chart
axes[1, 1].pie(sizes_alcohol, labels=labels_alcohol, autopct='%1.1f%%', colors=colors_alcohol, startangle=90)
axes[1, 1].set_title('Alcohol Consumption')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the title
plt.show()
```
